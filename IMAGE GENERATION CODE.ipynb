```python
## TRAIT GENERATION

TOTAL_INUS = 10000

traits = []

def createCombo():
    
    trait = {}

    trait["Background"] = random.choices(backgrounds, backgroundweights)[0]
    trait["Clothes"] = random.choices(clothes, clothesweights)[0]
    trait["Hats"] = random.choices(hats, hatweights)[0]
    trait["Mouth"] = random.choices(mouths, mouthweights)[0]
    trait["Fur"] = random.choices(fur, furweights)[0]
    trait["Eyes"] = random.choices(eyes, eyesweights)[0]
    trait["Earrings"] = random.choices(earrings, earringweights)[0]
    
    
    if trait in traits:
        return createCombo()
    else:
        return trait
    
for i in range(TOTAL_INUS):
    
    newtraitcombo = createCombo()
    
    traits.append(newtraitcombo)
    
```

```python
## VERIFY ALL IMAGES ARE UNIQUE
def allUnique(x):
    seen = list()
    return not any(i in seen or seen.append(i) for i in x)

print(allUnique(traits))
```
